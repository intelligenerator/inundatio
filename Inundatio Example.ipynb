```python
import glob
import numpy as np
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

from Inundatio import get_houses
```

```python
# Load the paths of the target images
images_dir = 'train/images'
images_list = sorted(glob.glob(images_dir + '/*_post_disaster.png'))
targets_dir = 'train/targets'
targets_list = sorted(glob.glob(targets_dir + '/*_post_disaster_target.png'))

# Load an image
image = Image.open(images_list[0])
plt.imshow(image, cmap='gray_r')
plt.show()

target_image = Image.open(targets_list[0])
target_image = np.array(target_image)
plt.imshow(target_image, cmap='gray_r')
plt.show()

# Overlay
plt.imshow(image)
plt.imshow(target_image, cmap='jet', alpha=0.3)
plt.show()
```

```python
coordinates_list = get_houses(target_image)
print(coordinates_list)
```
